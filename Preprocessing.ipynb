```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer

#reading the data
data=pd.read_csv(r"C:\Users\Dell\Desktop\dp\Data.csv")
data
```

```python
#checking for missing vakues gives total null entries in particular column 

data.isnull().sum()
```

```python
#filling out the missing values with the mean
data = data.fillna(method='ffill')
data
```

```python
#encoding for binary categorical data by using LabelBinarizer encoding occurs as yes=1 ,No=1

import sklearn.preprocessing as pp
lb=pp.LabelBinarizer()
data.Purchased=lb.fit_transform(data.Purchased)
data
```

```python

#onehot and label encoding for the categorical data here for Country column

data_Country=pd.get_dummies(data['Country'],prefix='Country')     #creating dummy variables
data_Country
```

```python

df2=data
del df2['Country']

df2= pd.concat([ data_Country,data],axis=1)   #axis= 1 means add column wise
df2.head(3)
```

```python
#clean data
df2
```

```python
#Splitting the data into dependent and independent variables

from sklearn.model_selection import train_test_split

dependent=df2.iloc[:,5].values
independent=df2.iloc[:,0:5].values

X=pd.DataFrame(independent, columns=['Country_France','Country_Germany','Country_Spain','Age','Salary'])  #independent
Y=pd.DataFrame(dependent,columns=['Purchased'])           #dependent
print(X) 
print(Y)
```

```python
#splitting data into training and testing set

from sklearn.model_selection import train_test_split as tt

X_train,Y_train,X_test,Y_test= tt(X,Y,test_size=0.2,random_state=0)

print('X_train:\n\n',X_train)
print('\nX_test:\n\n',X_test)
print('\nY_train:\n\n',Y_train)
print('\nY_test:\n\n',Y_test)

 
```
